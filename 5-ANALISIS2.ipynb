OPERACIONES CON PANDAS

            Convertir el signo de pesos que es string a floats

```python
preciosdf["Precio"]=preciosdf["Precio"].str.replace("$", "").astype("float")
```

```python
preciosdf["Costo"]=preciosdf["Costo"].str.replace("$", "").astype("float")
```

```python
preciosdf.head(5)
```

    Generar una fila de "Ganancias" = "Precio" - "Costo"

```python
preciosdf["Ganancias"]=preciosdf["Precio"]-preciosdf["Costo"]
preciosdf.head(5)
```

    Retirar una columna

```python
#DESASERTE DE UNA COLUMNA
restdf=restdf.drop("Hora de Cobro",axis=1)
restdf
```

    Generar un dataframe con los platillos de mas de 100 pesos

```python

carosdf=preciosdf[preciosdf["Precio"]>100]
carosdf.head(10)
```

    Generar un dataframe con los platillos que tienen mas ganancia

```python
#ARMAR UN DF CON LOS PLATILLOS QUE TIENEN GRANDES GANANCIAS
preciosdf["Margen"]=np.where(preciosdf["Ganancias"]>100,"Alto","Bajo")
preciosdf.head(10)
```

    Modificar el nombre de una columna para tener meseros

```python
#Cambiar un nombre de columna
restdf=restdf.rename(columns={"Atendió":"Mesero"})
```

```python
restdf.head(5)
```

BUSCARV VLOOKUP

    Realizando un left join (sql) con un merge

```python
restpreciodf=restdf.merge(preciosdf,on=["Producto"],how="left")
```

```python
restpreciodf.head(20)
```

TABLAS DINAMICAS

    Quitando Duplicados

```python
meseros=restdf["Mesero"]
meseros.head(5)
```

```python
meseros.drop_duplicates()
```

GENERAR TABLA DINAMICA CON PYTHON

    Tabla dinamica de ventas frente a categoria

```python
venxcat= pd.pivot_table(restpreciodf,values="Precio",index="Categoria", aggfunc=np.sum)
venxcat
```

    Tabla dinamica de ventas frente a categoria y tipo

Para armar una tabla dinamica que muestra valores por columna

```python
venxcatxtipo=pd.pivot_table(restpreciodf,values="Precio",index="Categoria",columns="Tipo_x",aggfunc=np.sum)
venxcatxtipo
```

    Tabla dinamica de ventas frente a categoria y tipo por cada mesero

```python
venxcatxtipoxmesero=pd.pivot_table(restpreciodf,values="Precio",index=["Tipo_x","Categoria"],columns="Mesero",aggfunc=np.sum)
venxcatxtipoxmesero
```

    Tabla dinamica de ventas promedio por mesero

```python
ventasxpromxmesero=pd.pivot_table(restpreciodf,values="Precio",index="Mesero")
ventasxpromxmesero
```
